```python
# To support both python 2 and python 3
from __future__ import division, print_function, unicode_literals

# Common imports
import numpy as np
import os

# to make this notebook's output stable across runs
np.random.seed(42)

# To plot pretty figures
%matplotlib inline
import matplotlib as mpl
import matplotlib.pyplot as plt
mpl.rc('axes', labelsize=14)
mpl.rc('xtick', labelsize=12)
mpl.rc('ytick', labelsize=12)

# Where to save the figures
PROJECT_ROOT_DIR = "."
CHAPTER_ID = "training_linear_models"

def save_fig(fig_id, tight_layout=True):
    path = os.path.join(PROJECT_ROOT_DIR, "images", CHAPTER_ID, fig_id + ".png")
    print("Saving figure", fig_id)
    if tight_layout:
        plt.tight_layout()
    plt.savefig(path, format='png', dpi=300)

# Ignore useless warnings (see SciPy issue #5998)
import warnings
warnings.filterwarnings(action="ignore", message="^internal gelsd")
```

```python
import numpy as np

X = 20 * np.random.rand(100, 1) - 10
y = 1 / (1 + np.exp(-X))
y_rand = 1 / (1 + np.exp(-X + 0.15 * np.random.randn(100, 1)))
```

```python
plt.plot(X, y, "r.")
plt.plot(X, y_rand, "b.")
plt.xlabel("$x_1$", fontsize=18)
plt.ylabel("$y$", rotation=0, fontsize=18)
t = np.linspace(-10, 10, 100)
sig = 1 / (1 + np.exp(-t))
plt.plot(t, sig, "k-", linewidth=2)
plt.axis([-10, 10, 0, 1])
plt.show()
```

```python
from sklearn.linear_model import LinearRegression
lin_reg = LinearRegression()
lin_reg.fit(X, y)
y_predict = lin_reg.predict(X)
lin_reg.intercept_, lin_reg.coef_
```

```python
from sklearn.metrics import mean_squared_error
mean_squared_error(y, y_predict)
```

```python
plt.plot(X, y_predict, "r.", linewidth=2, label="Predictions")
plt.plot(X, y, "b.")
plt.xlabel("$x_1$", fontsize=18)
plt.ylabel("$y$", rotation=0, fontsize=18)
plt.legend(loc="upper left", fontsize=14)
plt.axis([-10, 10, 0, 1])
plt.show()
```

```python
lin_reg_random = LinearRegression()
lin_reg_random.fit(X, y_rand)
y_random_predict = lin_reg_random.predict(X)
lin_reg_random.intercept_, lin_reg_random.coef_
```

```python
mean_squared_error(y_rand, y_random_predict)
```

```python
plt.plot(X, y_random_predict, "r.", linewidth=2, label="Predictions")
plt.plot(X, y, "b.")
plt.xlabel("$x_1$", fontsize=18)
plt.ylabel("$y$", rotation=0, fontsize=18)
plt.legend(loc="upper left", fontsize=14)
plt.axis([-10, 10, 0, 1])
plt.show()
```

```python
plt.plot(X, y_random_predict, "r.", linewidth=2, label="Predictions")
plt.plot(X, y_predict, "y.", linewidth=2, label="Random Predictions")
plt.plot(X, y, "b.")
plt.xlabel("$x_1$", fontsize=18)
plt.ylabel("$y$", rotation=0, fontsize=18)
plt.legend(loc="upper left", fontsize=14)
plt.axis([-10, 10, 0, 1])
plt.show()
```

```python
from sklearn.preprocessing import PolynomialFeatures
poly_features = PolynomialFeatures(degree=2, include_bias=False)
X_poly = poly_features.fit_transform(X)
```

```python
lin_reg_poly = LinearRegression()
lin_reg_poly.fit(X_poly, y)
y_poly_predict = lin_reg_poly.predict(X_poly)
lin_reg_poly.intercept_, lin_reg_poly.coef_
```

```python
plt.plot(X, y_poly_predict, "b.")
plt.plot(X, y, "k.")
plt.axis([-10, 10, 0, 1])
plt.show()
```

```python
from sklearn.preprocessing import PolynomialFeatures
def plotPolynomialRegression(deg, i):
    poly_features = PolynomialFeatures(degree=deg, include_bias=False)
    X_poly = poly_features.fit_transform(X)
    
    lin_reg_poly = LinearRegression()
    lin_reg_poly.fit(X_poly, y)
    y_poly_predict = lin_reg_poly.predict(X_poly)
    plt.subplot(2, 2, i)
    plt.plot(X, y_poly_predict, "b.")
    plt.plot(X, y, "k.")
    plt.axis([-10, 10, 0, 1])
    plt.set_figheight(15)
    plt.set_figwidth(15)
#     plt.show()
```

```python

plotPolynomialRegression(2, 1)
plotPolynomialRegression(3, 2)
plotPolynomialRegression(4, 3)
plotPolynomialRegression(5, 4)
```
